```python
import pandas as pd

data = pd.read_csv('todays_data.csv')
df = pd.DataFrame(data)

print(df.head()) # Display the first few rows of the dataset

print(df.dtypes) # Checks the data types

print(df.isnull().values.any()) # Checks any missing values

df.fillna(0) # Fills NaN values with 0

numerical_stats = df.describe() # Compute basic statistics of columns
print(numerical_stats)

AL_grouped = df.groupby('Academic Level').size()  # Group by 'academic level' and calculate the count of occurrences
print("\nGrouped by Academic Level:\n", AL_grouped)

import matplotlib.pyplot as plt

visual = df.value_counts('Academic Level') # Create Line chart of 'Academic Level' count
visual.plot()
plt.xlabel ('Academic Level')
plt.ylabel('Count')
plt.title('Academic Level vs Count')

bar = df.value_counts('Gender') # Create bar chart of 'Gender' count
bar.plot.bar()
plt.xlabel ('Gender')
plt.ylabel('Count')
plt.title('Gender vs Count')

histogram = df.value_counts('Country') # Create histogram chart of 'Country' count
histogram.plot.hist()
plt.xlabel('Country')
plt.title('Country Distribution')
```
